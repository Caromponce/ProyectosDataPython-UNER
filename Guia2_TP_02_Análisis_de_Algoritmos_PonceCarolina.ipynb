1) Factorial Recursivo: Escribe un programa en Python que calcule el factorial de un número entero utilizando una función recursiva. ¿Cuántas llamadas recursivas se realizan?

```python
# Resolución ejercicio 1:
numero = int(input("Ingrese un número entero: "))

def factorial(numero):
    if numero == 0:
        return 1
    else:
        return numero * factorial(numero - 1)

# Calcular el factorial y mostrar el resultado
resultado = factorial(numero)
print(f"El factorial de {numero} es {resultado}")
```

2) Escribir una función `buscar_ultimo_elemento()`que reciba una lista y un elemento y devuelva la posición de la última aparición de ese elemento en la lista o `-1` si el elemento no está en la lista.

```python
# Resolución ejercicio 2:
def crearLista():
  lista_numeros = []
  longitud_lista = int(input('Ingrese la cantidad de números que desea tener en la lista: '))
  for i in range(longitud_lista):
    lista_numeros.append(random.randint(0, 20))
  return lista_numeros

def buscar_ultimo_elemento(lista):
  numero_buscado = int(input('Ingrese el número entero que desea buscar: '))
  posicion = -1
  for i in range(len(lista)):
      if lista[i] == numero_buscado:
          posicion = i
  return numero_buscado, posicion

lista_creada = crearLista()
numero_buscado, num_posicion = buscar_ultimo_elemento(lista_creada)
print(f'La lista creada es {lista_creada}, el número buscado es {numero_buscado} y su última posición en la lista es {num_posicion}.')
```

```python
# Resolución ejercicio 4:
lista3 = [4, 9, 5, 1, 25, 15, 3, 22]
def maximo(lista):
  valor_maximo = lista[0]
  for item in lista:
    if valor_maximo < item:
      valor_maximo = item
  return valor_maximo

resultado = maximo(lista3)
print(f'El valor máximo de la lista {lista3} es {resultado}.')
```

5) Búsqueda Binaria: Implementa el algoritmo de búsqueda binaria en su versión recursiva. Dado un número objetivo y una lista ordenada de números, el programa debe determinar si el número objetivo está presente en la lista.

```python
# Resolución ejercicio 5:
def busqueda_binaria(lista, numero_buscado):
    izq = 0
    der = len(lista)-1
    medio = (izq + der)//2
    if len(lista) == 0:
      return -1
    if lista[medio] == numero_buscado:
        return medio
    else:
        if lista[medio] < numero_buscado:
            izq = medio + 1
        else:
            der = medio - 1
        lista_recortada = lista[izq:der]
        return busqueda_binaria(lista_recortada, numero_buscado)

def esta_presente(numero_buscado, resultado, lista4):
  if resultado == -1:
    return f'El numero buscado {numero_buscado} no está en la lista {lista4}'
  else:
    return f'El numero buscado {numero_buscado} si está en la lista {lista4}'

lista4 = [0, 1, 2, 8, 13, 17, 19, 32, 42]
numero_buscado = int(input('Ingrese el número que está buscando: '))
resultado = busqueda_binaria(lista4, numero_buscado)
print(esta_presente(numero_buscado, resultado, lista4))
```

6) Ordenamiento de Selección: Escribe una función que realice el ordenamiento de selección en una lista de números enteros.

```python
# Resolución ejericio 6:
def ordenamientoSeleccion(lista_desordenada):
  lista = lista_desordenada.copy()
  for i in range(len(lista)):
    min_index = i
    for j in range(i + 1, len(lista)):
      if lista[j] < lista[min_index]:
        min_index = j
    lista[i], lista[min_index] = lista[min_index], lista[i]
  return lista


lista_desordenada = [12, 45, 7, 23, 56, 19]
lista_ordenada = ordenamientoSeleccion(lista_desordenada)
print(f'La lista desordenada es {lista_desordenada} y la lista ordenada por selección es {lista_ordenada}.')
```

9) Algoritmo K-Means: El algoritmo K-Means se encuentra disponible en la biblioteca scikit-learn. Aplica el algoritmo a un conjunto de datos y muestra los grupos resultantes en un gráfico.

```python
# Resolución ejercicio 9:
data = list(zip(array_budget, array_revenue))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,n_init=10, random_state=0)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2,n_init=10)
kmeans.fit(data)

plt.scatter(array_budget, array_revenue,c=kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,n_init=10)
kmeans.fit(data)

plt.scatter(array_budget, array_revenue,c=kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,n_init=10)
kmeans.fit(data)

plt.scatter(array_budget, array_revenue,c=kmeans.labels_)
plt.show()
```
